```python
number=10
for i in range(number):
  rotated_img=cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
  cv2.imwrite(f'apple{i*1}.jpg',rotated_img)
  img=rotated_img.copy()
```

```python
number=5
for i in range(number):
  rotated_img=cv2.rotate(img,cv2.ROTATE_90_COUNTERCLOCKWISE)
  rotated_image=cv2.cvtColor(rotated_img,cv2.COLOR_BGR2GRAY)
  cv2.imwrite(f'applew{i}.jpg',rotated_image)
  img=rotated_image.copy()
```

```python
height,width=img.shape[:2]
center=(width/2,height/2)
angle_to_rotate=45
rotation_matrix=cv2.getRotationMatrix2D(center,angle_to_rotate,1.0)
rotated_image=cv2.warpAffine(img,rotation_matrix,(width,height))
plt.imshow(img)
plt.imshow(rotated_image)
```

```python
number=10
for i in range(number):
  height,width=img.shape[:2]
  center=(width/2,height/2)
  angle_to_rotate=random.uniform(0,360)
  rotation_matrix=cv2.getRotationMatrix2D(center,angle_to_rotate,1.0)
  rotated_image=cv2.warpAffine(img,rotation_matrix,(width,height))
  cv2.imwrite(f'applesweet{i*1}.jpg',rotated_image)
  img=rotated_image.copy()
```
